```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = open("data_train_mlr_200.csv", 'r')
names =["Area","Jumlah_Penghuni","Jumlah_Alat_Listrik","Jam_Penggunaan","Konsumsi_Listrik"]
dataset= pd.read_csv(data, names=names)
dataset
```

```python
dataset.describe()
```

```python
#Age
dataset.plot(x="Area", y="Konsumsi_Listrik", style="o")
plt.title("Pengaruh Area terhadap Konsumsi Listrik")
plt.xlabel("Pengaruh Area")
plt.ylabel("Konsumsi Listrik")
plt.show()
```

```python
#Bilirubin
dataset.plot(x="Jumlah_Penghuni", y="Konsumsi_Listrik", style="o")
plt.title("Pengaruh Jumlah Penghuni terhadap Konsumsi Listrik")
plt.xlabel("Jumlah Penghuni")
plt.ylabel("Konsumsi Listrik")
plt.show()
```

```python
#Alkaline Phosphatase
dataset.plot(x="Jumlah_Alat_Listrik", y="Konsumsi_Listrik", style="o")
plt.title("Pengaruh Jumlah_Alat Listrik terhadap Konsumsi Listrik")
plt.xlabel("Jumlah_Alat Listrik")
plt.ylabel("Konsumsi Listrik")
plt.show()
```

```python
#Albumin
dataset.plot(x="Jam_Penggunaan", y="Konsumsi_Listrik", style="o")
plt.title("Pengaruh Jam Penggunaan terhadap Konsumsi Listrik")
plt.xlabel("Jam Penggunaan")
plt.ylabel("Konsumsi Listrik")
plt.show()
```

```python
# Plot pertama
sns.jointplot(x='Area', y='Konsumsi_Listrik', data=dataset, kind='reg')

# Plot kedua
sns.jointplot(x='Jumlah_Penghuni', y='Konsumsi_Listrik', data=dataset, kind='reg')

# Plot ketiga
sns.jointplot(x='Jumlah_Alat_Listrik', y='Konsumsi_Listrik', data=dataset, kind='reg')

# Plot keempat
sns.jointplot(x='Jam_Penggunaan', y='Konsumsi_Listrik', data=dataset, kind='reg')

plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.pairplot(data=dataset, x_vars=['Area', 'Jumlah_Penghuni', 'Jumlah_Alat_Listrik', 'Jam_Penggunaan'], y_vars=['Konsumsi_Listrik'])
plt.show()
```
