```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = open("belajar2.csv", 'r')
names =["minat","kehadiran","prestasi"]
dataset= pd.read_csv(data, names=names)
dataset
```

```python
dataset.describe()
```

```python
#minat
dataset.plot(x="minat", y="prestasi", style="o")
plt.title("Pengaruh Minat Belajar pada Prestasi Belajar")
plt.xlabel("Minat Belajar")
plt.ylabel("Prestasi Belajar")
plt.show()
```

```python
#kehadiran
dataset.plot(x="kehadiran", y="prestasi", style="o")
plt.title("Pengaruh Kehadiran Belajar pada Prestasi Belajar")
plt.xlabel("kehadirann")
plt.ylabel("Prestasi Belajar")
plt.show()
```

```python
# Plot pertama
sns.jointplot(x='minat', y='prestasi', data=dataset, kind='reg')

# Plot kedua
sns.jointplot(x='kehadiran', y='prestasi', data=dataset, kind='reg')

plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.pairplot(data=dataset, x_vars=['minat', 'kehadiran'], y_vars=['prestasi'])
plt.show()
```
